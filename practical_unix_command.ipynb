# 커맨드라인 사용법

## 용어정리

`유닉스`
- 1970년대 초반에 개발된 운영체제
- 소프트웨어 개발 실행 가능한 플랫폼
- 쉽게 수정해서 다른 컴퓨터에 적용 가능
- 유닉스를 기반으로 하는 다양한 운영 체제 탄생

`CLI`
- Command Line Interface의 준말로, 말 그대로 커맨드를 통해 컴퓨터를 사용하는 방식
    - 필요한 라이브러리 설치 관리
    - 협업시 코드 관리
    - 웹 개발 환경 설정/서버 실행
    
`GUI`
- Graphic User Interface의 준말로, 화면과 마우스를 통해 컴퓨터를 사용하는 방식
    - 윈도우의 클릭 -> 파일 열기, 드래그 -> 파일 이동과 같은 것들. 

`프롬프트`
- 용자@기기명 ~(working_dir:현재 파일 위치) %/$(끝)

`터미널`
- 입력 출력을 보여주는 검은 창

`쉘/bash`
- bash는 쉘 중 특정 하나를 말하는 것이며, 인간의 커맨드를 컴퓨터가 알아들을 수 있게 변환해주는 역할 수행
    -  ubuntu : 사용하는 것이 **bash**
    -  mac : 쭉 사용해오다가, 특정 버전 이후로 **zsh** 를 사용한다.

## 터미널 사용하기

`아규먼트(인자)`
- 커맨드 뒤에는 `아규먼트(인자)`가 들어감
- 어느 대상에 대해 실행할 것인지 정함

`옵션`
- 커맨드 뒤에는 `옵션`이라는 것도 넣을 수 있음
- 옵션은 **-(대시)** 를 붙여서 보통 입력
- 커맨드를 어떻게 실행할 것인지 정해줌

간단한 단축키
1. `방향키 아래` : 한 줄씩 읽기
2. `f(forward)/space bar` : 다음 페이지 <-> `b(back)` : 이전 페이지
3. `q(quit)` : 누르면 다시 바깥 페이지로 돌아감


### 터미널 사용 꿀팁
- 이전에 사용했던 커맨드 기록 불러오기 : `위쪽 방향키`
- 커맨드 수정하기
    - 단어 단위 이동(`option+방향키`/`alt+방향키`)
    - 맨 앞/ 뒤: `ctrl+a`/ `ctrl+e`
- 취소하기 : `ctrl+c`
    - 커맨드 실행 전 : 그냥 새로운 프롬프트 시작
    - 커맨드 실행 후 : 실행 중단 및 새로운 프롬프트 시작
- 터미널 청소하기 : `clear 입력`
- 명령어 자동완성/목록 보여주기 : `tap 누르기`
    - 폴더 경로 입력시에도 편리하게 사용 가능 

# 파일과 폴더 다루기

## 폴더 구조 파일 경로

- 홈 디렉토리
    - 사용자 관련 (보통 우리에게 작업하는) 파일들이 모두 모여있는 곳
        1. 각종 문서
        2. 바탕화면 파일
        3. 다운로드 파일
        
    - 그밖의 것들은 컴퓨터에 일반적으로 필요한 환경 설정 파일들    

- `/` : root 디렉토리

- 현재 사용자의 홈 디렉토리: `~`
    - Ubuntu : /home/, MacOS : /Users/
    1. codeit 사용자 : /home/codeit
    2. taeho 사용자 : /home/taeho
   - 경로를 줄여서 사용 가능 : ~/Documents or ~/Documents/Projects


- `pwd` : print working directory의 준말
    - working directory : 현재 작업하고, 위치하고 있는 파일 경로


- `cd` : change directory의 준말
    - 경로 이동, 인자로 준 경로로 위치를 이동한다.
    - `cd (인자가 없는 경우)`: 홈 디렉토리로 이동
    - `cd -` : 이전 기록 디렉토리로 이동

- `ls` : list의 준말
    - 디렉토리의 내용들을 쭉 나열해서 목록으로 보여준다.
    - `ls (경로가 없는 경우)` : 현재 디렉토리의 목록을 보여준다.
    - `ls (특정 파일 경로)` : 특정 파일 경로의 목록을 보여준다.
    - <옵션>
        1. `-a` : **숨겨진 파일(확장자가 없다)까지** 모두 보여준다.

            - 확장자가 없다는 것이 무엇일까?
                * 사실 확장자는 일반 파일도 없을 수 있다. 가령, 텍스트
                    * 확장자는 파일의 형식, 어떤 프로그램을 통해 열 수 있는지에 대한 안내일 뿐이다.

        2. `-l` : long format의 준말로, **파일의 자세한 정보를 모두** 출력한다.
            * 파일의 종류, 권한, 링크의 개수, 소유자, 그룹, 바이트(용량X), 마지막 수정 날짜 시간, 이름 등

## 절대 / 상대 경로


- `<절대경로 vs 상대경로>`
    1. 절대 경로 : 루트 디렉토리를 기준으로 모든 경로를 작성하는 것.
    2. 상대 경로 : 현재 위치해있는 워킹 디렉토리를 기준으로 경로를 나타내는 것.
        - 현재 디렉토리를 `.`으로 표시한다.
